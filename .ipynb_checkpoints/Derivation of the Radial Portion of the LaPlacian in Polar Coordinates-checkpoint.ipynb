Keep in mind that the double factorial notation is defined to be the product of all even integers less than or equal to $x \geq 2$.  For an odd integer p, the double factorial is the product of all odd integers $\leq p$ and $\geq 1$.  The following are defined:
$$ 0!! = -1!! = 1$$
$$(x+1)!! = (x+1) \cdot (x-1) \cdot (x-3) \cdot (x-5) \ldots$$
So,
$$(2x+1)! = (2x+1)!! \cdot 2^x \cdot x!$$

Look at the patterns for the even indices $a_{2l}$.
$$a_{2l} = \frac{- a_{2l-2}}{(2l) \cdot (2l-1)}$$
$$a_{2} = \frac{- a_0}{(2l) \cdot (2l-2)}$$
\begin{split}
a_{4} & = \frac{- a_{2}}{2 \cdot (l-1) \cdot (2(l-1) -1)}
\\ & =
\frac{- a_{2}}{2 \cdot 2 \cdot (l-1) \cdot (2l-3) }
\\ & =
\frac{(-1)(-1) a_{0}}{(2l) \cdot (2l-1) \cdot 2 \cdot (l-1) \cdot (2l-3) }
\\ & =
\frac{a_{0}}{2^2 \cdot (l) \cdot (l-1) \cdot (2l-1) \cdot (2l-3)}
\end{split}

\begin{split}
a_{6} & = \frac{- a_{4}}{2 \cdot (l-2) \cdot (2(l-2) -1)}
\\ & =
\frac{- a_{4}}{2 \cdot (l-2) \cdot (2l-5) }
\\ & =
\frac{- a_{0}}{(2l) \cdot (l-2) \cdot (2l-5) \left[2^2 \cdot (l) \cdot (l-1) \cdot (2l-1) \cdot (2l-3)\right]}
\\ & =
\frac{-a_{0}}{2^3 \cdot (l) \cdot (l-1) \cdot (l-2) \cdot (2l-1) \cdot (2l-3) \cdot (2l-5) }
\end{split}


\begin{split}
\therefore a_{2l} & = \frac{(-1)^l \cdot a_0}{2^l \cdot (l!) \cdot \left[(2l-1) \cdot (2l-3) \cdot (2l-5) \right]}
\\ & =
\frac{(-1)^l \cdot a_0}{2^l \cdot (l!) \cdot (2l-1)!!}
\\ & =
\frac{(-1)^l \cdot a_0}{(2l)!}
\end{split}


Look at the patterns for the odd indices, or $a_{2l+1}$.  
$$a_{2l+1} = \frac{- a_{2l-1}}{(2l) \cdot (2l+1)}$$
$$a_{2l-3} = \frac{- a_{2l-5}}{2 \cdot (l-2) \cdot (2l-3)}$$
\begin{split}
a_{2l-5} & = \frac{- a_{2l-7}}{2 \cdot (l-3) \cdot (2l-5)}
\\ & =
\frac{- a_{2l-5}}{2 \cdot 2 \cdot (l-2) \cdot (l-3) \cdot (2l-3) \cdot (2l-5)}
\end{split}
\begin{split}
a_{2l-1} & = \frac{- a_{2l-3}}{2 \cdot (l-1) \cdot (2(l-1)+1)}
\\ & =
\frac{- a_{2l-5}}{2 \cdot (l-1) \cdot (2l-1) \cdot 2 \cdot (l-2) \cdot (2l-3)}
\end{split}

\begin{equation}
\begin{split}
a_{2l+1} & = \frac{- a_{2l-1}}{(2l) \cdot (2l+1)}
\\ & =
\frac{(-1)(-1) a_{2l-3}}{2l \cdot (2l+1) \cdot 2 \cdot (l-1) \cdot (2l-1)}
\\ & =
\frac{a_{2l-3}}{2^2 \cdot (l) \cdot (l-1) \cdot (2l+1) \cdot (2l-1)}
\\ & =
\frac{- a_{2l-5}}{2^2 \cdot (l) \cdot (l-1) \cdot (2l+1) \cdot (2l-1) \cdot (2) \cdot (l-2) \cdot (2l-3)}
\\ & =
\frac{- a_{2l-5}}{2^3 \cdot (l) \cdot (l-1) \cdot (l-2) \cdot (2l+1) \cdot (2l-1) \cdot (2l-3)}
\\ & =
\frac{(-1)(-1) a_{2l-7}}{2^3 \cdot (l) \cdot (l-1) \cdot (l-2) \cdot (2l+1) \cdot (2l-1) \cdot (2l-3) \cdot (2) \cdot (l-3) \cdot (2l-5)}
\\ & =
\frac{a_{2l-7}}{2^4 \cdot (l) \cdot (l-1) \cdot (l-2) \cdot (l-3) \cdot (2l+1) \cdot (2l-1) \cdot (2l-3) \cdot (2l-5)}
\\ & =
\frac{(-1)^l}{(2^l) \cdot l! \cdot (2l+1)!!}
\end{split}
\end{equation}

Since $(2n+1)! = (2n+1)!! \cdot 2^n \cdot n!$
$$a_{2l+1} = \frac{(-1)^l a_1}{(2l+1)!}$$



Summary:
\begin{equation}
\begin{split}
R(\rho) & = (1 + 2 \alpha)a_1 r^{\alpha + 1}
\\ & +
\sum_{n=2}^{\infty} \left[n(n+2\alpha)a_n + a_{n-2} \right] \cdot r^{\alpha + n}
\\ & = 0
\end{split}
\end{equation}

For this to be true all terms need to be zero.  This means that $\alpha = -\frac{1}{2}$.  If we substitute $k = \alpha$ then
\begin{equation}
\begin{split}
R(\rho) & = \sum_{n=0}^{\infty} a_n \rho^{k + n}
\\ & = 
\rho^{-\frac{1}{2}}\sum_{n=0}^{\infty} a_n \rho^{n}
\end{split}
\end{equation}


\begin{equation}
\begin{split}
R(\rho) & = \rho^{-\frac{1}{2}} \sum_{n=0}^{\infty}a_n \cdot \rho^n
\\ & =
\rho^{-\frac{1}{2}} \left[\sum_{n=1,3,5,\ldots}^{\infty}a_n \cdot \rho^n + \sum_{n=0,2,4,\ldots}^{\infty}a_n \cdot \rho^n  \right]
\\ & =
\rho^{-\frac{1}{2}} \left[\sum_{l=0}^{\infty}a_{2l} \cdot \rho^{2l} + \sum_{l=0}^{\infty}a_{2l+1} \cdot \rho^{2l+1}  \right]
\\ & =
\rho^{-\frac{1}{2}} \left[a_0 \sum_{l=0}^{\infty}\frac{-1^l}{(2l)!} \cdot \rho^{2l} + a_1\sum_{l=0}^{\infty}\frac{-1^l}{(2l+1)!} \cdot \rho^{2l+1}  \right]
\end{split}
\end{equation}
$$\sin(x) = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n+1}}{(2n+1)!}$$
$$\cos(x) = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n}}{(2n)!}$$
$$\therefore$$
$$R(\rho) = \rho^{-\frac{1}{2}} \left( a_0 \cos(\rho) + a_1 \sin(\rho) \right)$$
Bessel functions of order $\pm \frac{1}{2}$
$$J_{-\frac{1}{2}} (\rho) = \sqrt{\frac{2}{\pi \cdot \rho}} \cos(\rho)$$
$$J_{\frac{1}{2}} (\rho) = \sqrt{\frac{2}{\pi \cdot \rho}} \sin(\rho)$$
$$\therefore$$
$$R(\rho) = a'_0 J_{-\frac{1}{2}} (\rho) + a'_1 J_{\frac{1}{2}} (\rho)$$
