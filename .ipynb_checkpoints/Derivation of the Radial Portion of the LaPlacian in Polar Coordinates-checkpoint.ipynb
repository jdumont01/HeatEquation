Objective:
    Derive the radial portion of the Heat Equation in Polar coordinates.
    

We saw the radial components of partial differential equation include a $\frac{1}{\rho}$ term. This means that we have be mindful of how the center of the circle ($\rho = 0$) affects the overall solution since as $\rho \rightarrow 0$ the solution will approach $\infty$.  Conveniently, the Frobenius method for solving these types of differential equations can be employed because it includes a displacement factor around the singularity.

\begin{equation}
\begin{split} 
\rho^2  \frac {d^2 R(\rho)}{d \rho^2} + \rho \frac{d R(\rho)}{d \rho} + (\rho^2 \lambda - m^2)R(\rho) &= 0
\end{split}
\end{equation}

Start with the Frobenius equation:
\begin{equation}
\begin{split}
R(\rho) & = \sum_{n=0}^{\infty} a_n \rho^{\alpha + n}
\end{split}
\end{equation}

Subtituting this into the ODE
\begin{equation}
\begin{split}
R(\rho) &= \rho^2 \sum_{n=0}^{\infty} a_{n} (\alpha + n) (\alpha + n - 1) \rho^{\alpha + n - 2} + \rho \sum_{n=0}^{\infty} a_{n} (\alpha + n) \rho^{\alpha + n - 1} + (\rho^2 \lambda - m^2)\sum_{n=0}^{\infty} a_n \rho^{\alpha + n} \\
&= \rho^2 \sum_{n=0}^{\infty} a_n (\alpha + n) (\alpha + n - 1) \rho^{\alpha + n - 2} + \rho \sum_{n=0}^{\infty} a_{n} (\alpha + n) \rho^{\alpha + n - 1} + (\rho^2 \lambda - m^2)\sum_{n=0}^{\infty} a_n \rho^{\alpha + n} \\
&= \sum_{n=0}^{\infty} a_{n} (\alpha + n) (\alpha + n - 1) \rho^{\alpha + n} + \sum_{n=0}^{\infty} a_{n} (\alpha + n) \rho^{\alpha + n} + (\rho^2 \lambda - m^2)\sum_{n=0}^{\infty} a_n \rho^{\alpha + n} \\
&= \sum_{n=0}^{\infty} a_{n} (\alpha + n) (\alpha + n - 1) \rho^{\alpha + n} + \sum_{n=0}^{\infty} a_{n} (\alpha + n) \rho^{\alpha + n} + (\rho^2 \lambda - m^2)\sum_{n=0}^{\infty} a_n \rho^{\alpha + n} \\
&= \sum_{n=0}^{\infty} a_{n} (\alpha + n) (\alpha + n - 1) \rho^{\alpha + n} + \sum_{n=0}^{\infty} a_{n} (\alpha + n) \rho^{\alpha + n} + \lambda \sum_{n=0}^{\infty}{a_n \rho^{\alpha + n + 2}} - m^2 \sum_{n=0}^{\infty}{a_n \rho^{\alpha + n}}\\
&= \sum_{n=0}^{\infty} a_{n} (\alpha + n) (\alpha + n - 1) \rho^{\alpha + n} + \sum_{n=0}^{\infty} a_{n} (\alpha + n) \rho^{\alpha + n} + \lambda \sum_{n=2}^{\infty}{a_{n-2} \rho^{\alpha + n}} - m^2 \sum_{n=0}^{\infty}{a_n \rho^{\alpha + n}}\\
&= \sum_{n=0}^{\infty} \left[ a_{n} (\alpha + n) (\alpha + n - 1) \rho^{\alpha + n} +
a_{n} (\alpha + n) \rho^{\alpha + n} - m^2 a_n \rho^{\alpha + n} \right] + \lambda \sum_{n=2}^{\infty}{a_{n-2} \rho^{\alpha + n}} \\ 
&= \sum_{n=0}^{\infty} \left[ a_{n} (\alpha + n) (\alpha + n - 1)  +
a_{n} (\alpha + n) - m^2 a_n  \right] \rho^{\alpha + n} + \lambda \sum_{n=2}^{\infty}{a_{n-2} \rho^{\alpha + n}} = 0
\end{split}
\end{equation}

The next step will be to find the indicial equation from which, we will find a pattern (hopefully), to transform the infinite series into a more easily discernable equation.  As stated in equation 1, the differential equation is equal to 0.  In turn, this means that the sum of the coefficients of the sum must be zero.

To find the inidicial equation we set n=0.  The second term in (3) will not participate, until n=2.

\begin{equation}
\begin{split}
a_{0} (\alpha) (\alpha - 1)  + a_{0} \alpha - m^2 a_0 &= 0 \\
a_{0} \left((\alpha) (\alpha - 1)  + \alpha - m^2 \right) &= 0 \\
a_{0} \left(\alpha^{2} - m^2 \right) &= 0 \\
\end{split}
\end{equation}

The non-trivial solution will be when 

\begin{equation}
\begin{split}
\alpha^{2} &= m^{2} \\
\text{or} \\
\alpha &= \pm m
\end{split}
\end{equation}

If we take the positive m such that $\alpha = +m $ and substitute it back into (3) then
\begin{equation}
\begin{split}
\sum_{n=0}^{\infty} \left[ a_{n} (m + n) (m + n - 1)  +
a_{n} (m + n) - m^2 a_n  \right] \rho^{m + n} + \lambda \sum_{n=2}^{\infty}{a_{n-2} \rho^{m + n}} &= \\
\sum_{n=0}^{\infty} \left[ a_{n} n(n + 2m) \right] \rho^{m + n} + \lambda \sum_{n=2}^{\infty}{a_{n-2} \rho^{m + n}} &= \\
\text{since n=0 term leads to the left-hand side being zero, this sum can be updated to be}\\
\sum_{n=1}^{\infty} \left[ a_{n} n(n + 2m) \right] \rho^{m + n} + \lambda \sum_{n=2}^{\infty}{a_{n-2} \rho^{m + n}} &= 0
\end{split}
\end{equation}

Setting n=1 leads to the following
\begin{equation}
\begin{split}
\left[ a_{1} (2m + 1) \right] \rho^{m + 1} + \lambda \sum_{n=2}^{\infty}{a_{n-2\ \rho^{m + n}}} &= 0 \\
\text{the right-hand \ does not participate, so the left-hand term above must be zero} \\
\left[ a_{1} (2m + 1) \right] \rho^{m + 1}  &= 0  
\end{split}
\end{equation}
Since m is not zero, then $a_{1}$ = 0 .      
Now we can look at the n=2 ter
a_{n} n(n + 2m) &= - \lambda a_{n-2}
ms.
\begin{equation}
\begin{split}
\sum_{n=2}^{\infty}{\left[ a_{n} n(n + 2m) + \lambda a_{n-2} \right] \rho^{m + n}} &= 0
\end{split}
\end{equation}

Since the sum must be zero, then all of the coefficients within the sum must be zero.  
\begin{equation}
\begin{split}
a_{n} n(n + 2m) + \lambda a_{n-2} &= 0 \\
\lambda a_{n-2} &= -a_{n} n(n + 2m)  \\
a_{n-2} &= -\frac{n(n + 2m)}{\lambda} a_{n} \\
\text{since n=2} \\
a_{n-2} &= -\frac{n(n + 2m)}{\lambda} a_{n} 
\end{split}
\end{equation}


\begin{equation}
\begin{split}
 (1 + 2 \alpha)a_1 r^{\alpha + 1} \\
& +
\sum_{n=2}^{\infty} \left[n(n+2\alpha)a_n + a_{n-2} \right] \cdot r^{\alpha + n}
& = 0
\end{split}
\end{equation}

\begin{equation}
\begin{split}
 (1 + 2 \alpha)a_1 r^{\alpha + 1} \\
& +
\sum_{n=2}^{\infty} \left[n(n+2\alpha)a_n + a_{n-2} \right] \cdot r^{\alpha + n}
& = 0
\end{split}
\end{equation}

For this to be true all terms need to be zero.  This means that 
$$a_1 \cdot \left(1 + 2\alpha\right) = 0$$
and 
$$\left[n(n+2\alpha)a_n + a_{n-2} \right] = 0 $$
so
$$n(n+2\alpha)a_n = - a_{n-2}$$
or
$$a_n = - \frac{a_{n-2}}{n(n+2\alpha)}$$

Consider the case where $\alpha = -\frac{1}{2}$.  From the $a_1 \cdot \left(1 + 2\alpha\right) = 0$ term $a_1 = 0$, and by extension, all odd terms in the variable n are 0 due to the recursion relationship.  Consequently, because there is an indefinite number of even terms that remain in the sum, we can change the index from $n \longrightarrow 2l$ so that 
$$a_n = \frac{-a_{n-2}}{(n)(n-1)}$$
for n = 2,4,6,8, $\ldots$
becomes
$$a_{2l} = \frac{-a_{2l-2}}{(2l)(2l-1)}$$
for l = 1,2,3,4 $\ldots$


If we substitute $k = \alpha$ then
\begin{equation}
\begin{split}
R(\rho) & = \sum_{n=0}^{\infty} a_n \rho^{k + n}
\\ & = 
\rho^{-\frac{1}{2}}\sum_{n=0}^{\infty} a_n \rho^{n}
\end{split}
\end{equation}


Keep in mind that the double factorial notation is defined to be the product of all even integers less than or equal to $x \geq 2$.  For an odd integer p, the double factorial is the product of all odd integers $\leq p$ and $\geq 1$.  The following are defined:
$$ 0!! = -1!! = 1$$
$$(x+1)!! = (x+1) \cdot (x-1) \cdot (x-3) \cdot (x-5) \ldots$$
So,
$$(2x+1)! = (2x+1)!! \cdot 2^x \cdot x!$$

Look at the patterns for the even indices $a_{2l}$.
$$a_{2l} = \frac{- a_{2l-2}}{(2l) \cdot (2l-1)}$$
$$a_{2} = \frac{- a_0}{(2l) \cdot (2l-2)}$$
\begin{split}
a_{4} & = \frac{- a_{2}}{2 \cdot (l-1) \cdot (2(l-1) -1)}
\\ & =
\frac{- a_{2}}{2 \cdot 2 \cdot (l-1) \cdot (2l-3) }
\\ & =
\frac{(-1)(-1) a_{0}}{(2l) \cdot (2l-1) \cdot 2 \cdot (l-1) \cdot (2l-3) }
\\ & =
\frac{a_{0}}{2^2 \cdot (l) \cdot (l-1) \cdot (2l-1) \cdot (2l-3)}
\end{split}

\begin{split}
a_{6} & = \frac{- a_{4}}{2 \cdot (l-2) \cdot (2(l-2) -1)}
\\ & =
\frac{- a_{4}}{2 \cdot (l-2) \cdot (2l-5) }
\\ & =
\frac{- a_{0}}{(2l) \cdot (l-2) \cdot (2l-5) \left[2^2 \cdot (l) \cdot (l-1) \cdot (2l-1) \cdot (2l-3)\right]}
\\ & =
\frac{-a_{0}}{2^3 \cdot \left[(l) \cdot (l-1) \cdot (l-2) \right] \cdot (2l-1) \cdot (2l-3) \cdot (2l-5) }
\end{split}


\begin{split}
\therefore a_{2l} & = \frac{(-1)^l \cdot a_0}{2^l \left[\cdot (l!)\right] \cdot \left[(2l-1) \cdot (2l-3) \cdot (2l-5) \right]}
\\ & =
\frac{(-1)^l \cdot a_0}{2^l \cdot (l!) \cdot (2l-1)!!}
\\ & =
\frac{(-1)^l \cdot a_0}{(2l)!}
\end{split}


Look at the patterns for the odd indices, or $a_{2l+1}$.  
$$a_{2l+1} = \frac{- a_{2l-1}}{(2l) \cdot (2l+1)}$$
$$a_{2l-3} = \frac{- a_{2l-5}}{2 \cdot (l-2) \cdot (2l-3)}$$
\begin{split}
a_{2l-5} & = \frac{- a_{2l-7}}{2 \cdot (l-3) \cdot (2l-5)}
\\ & =
\frac{- a_{2l-5}}{2 \cdot 2 \cdot (l-2) \cdot (l-3) \cdot (2l-3) \cdot (2l-5)}
\end{split}
\begin{split}
a_{2l-1} & = \frac{- a_{2l-3}}{2 \cdot (l-1) \cdot (2(l-1)+1)}
\\ & =
\frac{- a_{2l-5}}{2 \cdot (l-1) \cdot (2l-1) \cdot 2 \cdot (l-2) \cdot (2l-3)}
\end{split}

\begin{equation}
\begin{split}
a_{2l+1} & = \frac{- a_{2l-1}}{(2l) \cdot (2l+1)}
\\ & =
\frac{(-1)(-1) a_{2l-3}}{2l \cdot (2l+1) \cdot 2 \cdot (l-1) \cdot (2l-1)}
\\ & =
\frac{a_{2l-3}}{2^2 \cdot (l) \cdot (l-1) \cdot (2l+1) \cdot (2l-1)}
\\ & =
\frac{- a_{2l-5}}{2^2 \cdot (l) \cdot (l-1) \cdot (2l+1) \cdot (2l-1) \cdot (2) \cdot (l-2) \cdot (2l-3)}
\\ & =
\frac{- a_{2l-5}}{2^3 \cdot (l) \cdot (l-1) \cdot (l-2) \cdot (2l+1) \cdot (2l-1) \cdot (2l-3)}
\\ & =
\frac{(-1)(-1) a_{2l-7}}{2^3 \cdot (l) \cdot (l-1) \cdot (l-2) \cdot (2l+1) \cdot (2l-1) \cdot (2l-3) \cdot (2) \cdot (l-3) \cdot (2l-5)}
\\ & =
\frac{a_{2l-7}}{2^4 \cdot (l) \cdot (l-1) \cdot (l-2) \cdot (l-3) \cdot (2l+1) \cdot (2l-1) \cdot (2l-3) \cdot (2l-5)}
\\ & =
\frac{(-1)^l}{(2^l) \cdot l! \cdot (2l+1)!!}
\end{split}
\end{equation}

Since $(2n+1)! = (2n+1)!! \cdot 2^n \cdot n!$
$$a_{2l+1} = \frac{(-1)^l a_1}{(2l+1)!}$$



\begin{equation}

\end{equation}
$$\sin(x) = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n+1}}{(2n+1)!}$$
$$\cos(x) = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n}}{(2n)!}$$
$$\therefore$$
$$R(\rho) = \rho^{-\frac{1}{2}} \left( a_0 \cos(\rho) + a_1 \sin(\rho) \right)$$
Bessel functions of order $\pm \frac{1}{2}$
$$J_{-\frac{1}{2}} (\rho) = \sqrt{\frac{2}{\pi \cdot \rho}} \cos(\rho)$$
$$J_{\frac{1}{2}} (\rho) = \sqrt{\frac{2}{\pi \cdot \rho}} \sin(\rho)$$
